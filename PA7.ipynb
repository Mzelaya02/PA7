```python
import math
class calculator():
    def __init__(self,input1,input2):
        self.input1=input1
        self.input2=input2
    def division (self):
        return self.input1/self.input2
    def multiplication(self):
        return self.input1*self.input2
    def subtraction(self):
        return self.input1-self.input2
    def addition(self):
        return self.input1+self.input2
    def squareroot(self):
        return math.sqrt(round(self.input1)),math.sqrt(round(self.input2))
    def exponential(self):
        return math.pow(self.input1,self.input2)
    def factorial(self):
        return math.factorial(round(self.input1)),math.factorial(round(self.input2))
    def integersquareroot(self):
        return math.isqrt(round(self.input1)),math.isqrt(round(self.input2))
    def roundup(self):
        return math.ceil(self.input1),math.ceil(self.input2)
    def rounddown(self):
        return math.floor(self.input1),math.floor(self.input2)
    def truncate(self):
        return math.trunc(self.input1),math.trunc(self.input2)
    def actual_rounding(self):
        return round(self.input1),round(self.input2)
    def absolute_value(self):
        return math.fabs(self.input1),math.fabs(self.input2)
    def permutation(self):
        return math.perm(round(self.input1),round(self.input2))
    def combination(self):
        return math.comb(round(self.input1),round(self.input2))
    def modulo(self):
        return self.input1 % self.input2
    
input_1=float(input("Enter first number: "))
input_2=float(input("Enter second number: "))
calc=calculator(input_1,input_2)

choice=1
while choice!=0:
    print()
    print("0. Exit")
    print("1. division")
    print("2. multiplication")
    print("3. subtraction")
    print("4. addition")
    print("5. squareroot")
    print("6. exponential")
    print("7. factorial")
    print("8. integersquareroot")
    print("9. roundup")
    print("10. rounddown")
    print("11. truncate")
    print("12. actual rounding")
    print("13. absolute value")
    print("14. permutation")
    print("15. combination")
    print("16. modulo")
    choice=int(input("Enter your choice..."))
    
    if choice==1:
        print(f"Result: {calc.division()}")
    
    elif choice==2:
        print(f"Result: {calc.multiplication()}")
    
    elif choice==3:
        print(f"Result: {calc.subtraction()}")
    
    elif choice==4:
        print(f"Result: {calc.addition()}")
   
    elif choice==5:
        print(f"Result: {calc.squareroot()}")
    
    elif choice==6:
        print(f"Result: {calc.exponential()}")
    
    elif choice==7:
        print(f"Result: {calc.factorial()}")
    
    elif choice==8:
        print(f"Result: {calc.integersquareroot()}")

    elif choice==9:
        print(f"Result: {calc.roundup()}")
    
    elif choice==10:
        print(f"Result: {calc.rounddown()}")
    
    elif choice==11:
        print(f"Result: {calc.truncate()}")
    
    elif choice==12:
        print(f"Result: {calc.actual_rounding()}")
    
    elif choice==13:
        print(f"Result: {calc.absolute_value()}")
    
    elif choice==14:
        print(f"Result: {calc.permutation()}")
    
    elif choice==15:
        print(f"Result: {calc.combination()}")
    
    elif choice==16:
        print(f"Result: {calc.modulo()}")

    elif choice==0:
        print("Exit")
        
    else:
        print("Sorry. Error.")
    
```
